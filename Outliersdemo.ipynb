```python
import numpy as np
```

```python
arr1 =[1,2,3,7,9,12,13,17,19,201]
arr2 =np.array(arr1)
```

```python
arr2
```

```python
Q1= int(np.percentile(arr2,25))
Q3= int(np.percentile(arr2,75))
```

```python
IQR=int(Q3-Q1)
IQR
```

```python
UF=int(Q3+1.5*IQR)
LF=int(Q1-1.5*IQR)
```

```python
l2=[]
```

```python
for i in arr2:
   if i>=LF and i<=UF:
       l2.append(i)
```
